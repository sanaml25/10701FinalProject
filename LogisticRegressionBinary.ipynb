```python
import pandas as pd 
import numpy as np 
import re
from nltk.corpus import stopwords
import pickle
```

```python
def ifelse(p, t, f):
    if p:
        return t
    return f
```

```python
file = open("train4.pkl", "rb")
train = pickle.load(file)
scores = list(map(lambda x: ifelse(x <= 4, 0, 1), train['y']))
X = train.drop(['y'], axis = 1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, scores, train_size = 0.75)
```

```python
lr = LogisticRegression(verbose = 1, solver = 'sag', max_iter = 100)
lr.fit(X_train, y_train)
```

```python
np.argmax((np.abs(lr.coef_)))
```
